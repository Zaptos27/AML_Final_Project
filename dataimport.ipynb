```python
import torch
import os
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
data_path = 'data/train'

data = torch.load(os.path.join(data_path, 'Track00001.pt'))
```

```python
lenght = data['mix'].shape[1]
C = 2
L  = 129
```

```python
torch.max(torch.randint(0, lenght, (100*lenght,)))
```

```python
data['mix'] = torch.view_as_real(data['mix'])
```

```python
amount = 100
```

```python
dat = torch.zeros(amount,L,2*C+1,2, dtype=torch.float64)
```

```python
data['mix'][:, 0:5].shape
```

```python
for j, rand in enumerate(torch.randint(0, lenght, (amount,))):
    if rand < C:
        dat[j][0] = torch.zeros(L, 2)
        i = 1
        while rand-C+i < 0:
            dat[j][i] = torch.zeros(L, 2)
            i+=1
        dat[j][i:] = data['mix'][:, rand-C+i:rand+C+1]
    elif rand > lenght-1-C:
        dat[j][-1] = torch.zeros(L, 2)
        i = 1
        while rand+C-i > lenght-1:
            dat[j][-1-i] = torch.zeros(L, 2)
            i+=1
        dat[j][:-1-i] = data['mix'][:, rand-C:rand+C+1-i]
    else:
        dat[j] = data['mix'][:, rand-C:rand+C+1]
```
